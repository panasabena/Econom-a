#### 1.1.Tasa de actividad, empleo, desocupación, subocupación, empleo no registrado según sexo
   * Tasa de actividad: Relación porcentual entre la población económicamente activa (PEA) y la poblción total.
   * Tasa de empleo: Relación porcentual entre la población ocupada y la población total.
   * Tasa de desocupación: Relación porcentual entre la población desocupada y la PEA.
   * Tasa de subocupación: Relación porcentual entre la población subocupada y la PEA.
   * Tasa de empleo no registrado: Relación porcentual entre los asalariados no registrados y el total de asalariados.
   * Desempleo: Situación de la persona que está en condiciones de trabajar pero no tiene empleo o lo ha perdido.
   * Desocupación: Se refiere a personas que, no teniendo ocupación, están buscando activamente trabajo.
   * Subocupación: Un subocupado es una persona que conforma la población económicamente activa, que no integra el colectivo de los desocupados pues tiene trabajo, pero de menos horas que las que se consideran necesarias para una jornada laboral completa, teniendo las condiciones físicas y de edad necesarias.
   * Empleo no registrado: es el trabajo popularmente llamado "trabajo en negro", que se da cuando el empleador no registra su relación laboral con un trabajador o cuando se la registra deficientemente (falseando por ejemplo, la fecha de ingreso, las horas trabajadas o la remuneración).

### Observaciones de los gráficos:
* Se aprecia en el primer gráfico (tasa de actividad en relación al tiempo), que durante el paso de los años, la brecha de actividad laboral entre ambos géneros se mantuvo en ± 13 %. Lo que indica, que por más que haya cambiado la oferta y demanda laboral, la brecha quedó intacta.
* Tasa de empleo: básicamente con las mismas características que el gráfico de actividad.
* Tasa de desocupación: Se puede observar que desde el año 2005 hasta el 2015, la desocupación venía en descenso, hasta algún acontecimiento o serie de acontecimientos en el 2016 que hizo que la desocupación aumentara. La brecha entre géneros en relación a la desocupación, se ha mantenido estrecha y constante durante el tiempo.
* Tasa de subocupación: este concepto nos indica que la persona trabaja, pero sigue teniendo disponibilidad horaria, aptitudes físicas y/o académicas. Se puede apreciar que el comportamiento es muy similar a la desocupación. La subocupación desciende desde 2005 hasta el 2015, luego a partir del 2016 empieza a incrementar.
* Tasa de empleo no registrado: Se aprecia en el gráfico, que la tasa de "Trabajo en negro", ha ido disminuyendo, hasta el 2012. Luego asciende levemente hasta el 2020.
